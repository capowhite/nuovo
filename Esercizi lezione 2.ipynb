```python
# version 1
def mod (a): 
    """ this function returns a true if the number is even , otherwise returns false"""
    result = a % 2 
    if result == 1 :
        return False 
    else: return True 

mod(11)
    
```

```python
# version 2
def mod2 (a): 
    """ this function returns a true if the number is even , otherwise returns false""" 
    if a % 2 == 1 :
        return False 
    else: return True 

mod2(10)
```

```python
# version 3 
def mod2 (a): 
    """ this function returns a true if the number is even , otherwise returns false""" 
    if a % 2 == 1 :
        return False 
    return True 

mod2(10)
```

```python
# version 4
def mod2 (a): 
    """ this function returns a true if the number is even , otherwise returns false""" 
    return ( a % 2 == 0) # this created a boolean value that return true of the result 
                         #is equal does not have decimals when divided, othersie returns FALSE

mod2(10)
```
